```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
orig = "Jakob/rhone-2/03-thinning/output-decr01.nc"
snap = "Jakob/rhone-2/06-fwd-from-snap/output-fwd-snap-from-withNNretrain.nc"
tcal = "Jakob/rhone-2/07-fwd-from-tcal/output-fwd-tcal-from-withNNretrain.nc"
```

```python
with xr.open_dataset(orig) as ds_orig:
    thk_orig = ds_orig["thk"].sum(dim=["x","y"]) * 2500. / 1.e9

with xr.open_dataset(snap) as ds_snap:
    thk_snap = ds_snap["thk"].sum(dim=["x","y"]) * 2500. / 1.e9

with xr.open_dataset(tcal) as ds_tcal:
    thk_tcal = ds_tcal["thk"].sum(dim=["x","y"]) * 2500. / 1.e9
```

```python
time_o = range(0,160,10)
time_inv = range(15,160,10)
```

```python
plt.plot(time_o,thk_orig,label='orig')
plt.plot(time_inv,thk_snap,label='snap')
plt.plot(time_inv,thk_tcal,label='tcal')

plt.legend()
plt.xlabel('time (yrs)')
plt.ylabel('volume (1e9 m^3)')
plt.show()
```
